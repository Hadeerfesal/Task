```python
import networkx as nx
```

```python
g= nx.Graph()
g.add_nodes_from(['a', 'b', 'c', 'd'])
g.add_edges_from([('a','b'), ('b','c'), ('c','d'), ('a','c')])
nx.draw(g, with_labels = True)
```

```python
list(g.neighbors('b'))
```

```python
for neighbor in g.neighbors('b'):
    print(neighbor)
```

```python
g.has_node('a')
```

```python
nx.is_tree(g)
```

```python
nx.is_connected(g)
```

```python
g.degree('a')
```

```python
Drawg= nx.DiGraph()
drawg.add_edges_from([('1','2'), ('2','3'), ('1','4'), ('3','4')])
nx.draw(drawg, with_labels = True)
```

```python
G=nx.Graph()
G.add_edges_from([(1,2),(2,3),(1,3),(3,4)])
nx.draw(G,with_labels=True,node_size=500)
```

```python
dict(G.degree())
```

```python
nx.to_numpy_matrix(G)
```
